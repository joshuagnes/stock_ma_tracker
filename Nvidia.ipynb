```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data generation
# Normally, we would fetch real stock data, but for this example, I'll create a simulated dataset
np.random.seed(42)
dates = pd.date_range(start="2023-01-01", periods=200)
prices = np.cumsum(np.random.randn(200) * 2 + 100)

# Create a DataFrame
data = pd.DataFrame(data={'Price': prices}, index=dates)

# Calculate 5-day and 10-day moving averages
data['5-Day MA'] = data['Price'].rolling(window=5).mean()
data['10-Day MA'] = data['Price'].rolling(window=10).mean()

# Plot the data
plt.figure(figsize=(14, 7))
plt.plot(data['Price'], label='Price', color='blue')
plt.plot(data['5-Day MA'], label='5-Day MA', color='orange')
plt.plot(data['10-Day MA'], label='10-Day MA', color='green')
plt.title('Stock Price with 5-Day and 10-Day Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()

data.tail()  # Show the last few rows of the data to get the latest moving average values
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download NVIDIA stock data
nvidia_data = yf.download("NVDA", start="2023-01-01", end="2024-06-17")

# Calculate 5-day and 10-day moving averages
nvidia_data['5-Day MA'] = nvidia_data['Close'].rolling(window=5).mean()
nvidia_data['10-Day MA'] = nvidia_data['Close'].rolling(window=10).mean()

# Plot the data
plt.figure(figsize=(14, 7))
plt.plot(nvidia_data['Close'], label='Close Price', color='blue')
plt.plot(nvidia_data['5-Day MA'], label='5-Day MA', color='orange')
plt.plot(nvidia_data['10-Day MA'], label='10-Day MA', color='green')
plt.title('NVIDIA Stock Price with 5-Day and 10-Day Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()

nvidia_data.tail(30)  # Show the last few rows of the data to get the latest moving average values
```
